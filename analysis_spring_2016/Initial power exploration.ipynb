```python
import scipy.interpolate as spi
import numpy as np
```

```python
def get_interpolated(x, y, xnew):
    fn = spi.interp1d(x=x, y=y, bounds_error=False, fill_value=x.iloc[-1])
    return fn(xnew)
```

```python
xnew = np.arange(first_ios_df.ts.iloc[0], first_ios_df.ts.iloc[-1], 30*60)
```

```python
first_ios_df_interpolated = get_interpolated(first_ios_df.ts, first_ios_df.battery_level_pct, xnew)
```

```python
iphone_drain_interpolated = map(lambda df: get_interpolated(df.ts, df.battery_level_pct, xnew), iphone_battery_df)
```
