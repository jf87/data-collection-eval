```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_for_transition(iloc, location_dfs, transition_dfs, axes_array, diff_axes_array, ground_truth_exit_ts, ground_truth_end_ts):
    end_ts_array = np.zeros(len(transition_dfs))
    for i, transition_df in enumerate(transition_dfs):
        if len(transition_df) > 0:
            # print transition_df[transition_df.transition == 2].tail()
            trip_end_ts = transition_df[(transition_df.transition == 2) | (transition_df.transition == 8)].iloc[iloc].ts
            end_ts_array[i] = trip_end_ts
        else:
            end_ts_array[i] = float("inf")

    min_end_ts = end_ts_array.min()
    print end_ts_array, "->", min_end_ts

    for i, df in enumerate(location_dfs):
        print "Considering phone %d" % i
        transition_df = transition_dfs[i]
        if len(transition_df) == 0:
            if len(df) == 0:
                print "No data to plot for dataframe %d" % i
                location_dfs[i-1][(location_dfs[i-1].ts > ground_truth_exit_ts) & (location_dfs[i-1].ts < min_end_ts)].plot(x="ts", y="latitude", kind="scatter", ax=axes_array[i])
                axes_array[i].annotate("No data for dataframe", xy=(0.3, 0.5), xycoords='axes fraction', fontsize=12, color="red")
            else:
                df[(df.ts > ground_truth_exit_ts) & (df.ts < min_end_ts)].plot(x="ts", y="latitude", kind="scatter", ax=axes_array[i])
                axes_array[i].axvline(x=ground_truth_exit_ts, linewidth=1, color="purple")
        else:
            trip_geofence_exit_entry = transition_df[transition_df.transition == 1].iloc[iloc]
            trip_geofence_exit_ts = trip_geofence_exit_entry["ts"]
            if trip_geofence_exit_ts <= ground_truth_exit_ts:
                print "trip_geofence_exit_ts = %d, ground_truth_exit_ts = %d, geofence was not exited!!" % (trip_geofence_exit_ts, ground_truth_exit_ts)
                location_dfs[i-1][(location_dfs[i-1].ts > ground_truth_exit_ts) & (location_dfs[i-1].ts < min_end_ts)].plot(x="ts", y="latitude", kind="scatter", ax=axes_array[i])
                axes_array[i].annotate("Geofence was not exited!", xy=(0.3, 0.5), xycoords='axes fraction', fontsize=12, color="red")
            else:
                df[(df.ts > trip_geofence_exit_ts) & (df.ts < ground_truth_end_ts)].plot(x="ts", y="latitude", kind="scatter", ax=axes_array[i])
                first_location_entry = df[(df.ts > trip_geofence_exit_ts) & (df.ts < ground_truth_end_ts)].iloc[0]
                print "geofence_exit_ts = %s, first location point = %s, delta = %5f" % \
                    (trip_geofence_exit_entry.fmt_time, first_location_entry.fmt_time, first_location_entry.ts - trip_geofence_exit_entry.ts)
                diff_df = df[(df.ts > trip_geofence_exit_ts) & (df.ts < ground_truth_end_ts)].ts.diff()
                print diff_df[diff_df > 30].head()
                diff_df.plot(ax=diff_axes_array[i])
                axes_array[i].axvline(x=trip_geofence_exit_ts, linewidth=1, color="green")
                axes_array[i].axvline(x=ground_truth_exit_ts, linewidth=1, color="purple")
                # axes_array[i].axvline(x=min_end_ts, linewidth=1, color="red")
```

```python
import emission.core.wrapper.transition as ecwt
```

```python
def insert_transition_strings(df_list):
    for df in df_list:
        if len(df) > 0:
            df["transition_str"] = df.transition.map(lambda(tval): ecwt.TransitionType(tval))
```
