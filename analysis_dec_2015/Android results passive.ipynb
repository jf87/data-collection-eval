```python
import numpy as np

fig,ax = plt.subplots(nrows=3, ncols=2, figsize=(15,6), sharey=True)
regimes = ['noha30s', 'nohanf']
for i, regime in enumerate(regimes):
    filtered_df = phone1_power_results[(phone1_power_results.stat=='drop_gauge') & (phone1_power_results.regime == regime)]
    if len(filtered_df) != 0:
        filtered_df.plot(ax=ax[0][i])
        
for i, regime in enumerate(regimes):
    filtered_df = phone2_power_results[(phone2_power_results.stat=='drop_gauge') & (phone2_power_results.regime == regime)]
    filtered_df.plot(ax=ax[1][i])
        
for i, regime in enumerate(regimes):
    filtered_df = phone2_power_results[(phone3_power_results.stat=='drop_gauge') & (phone3_power_results.regime == regime)]
    filtered_df.plot(ax=ax[2][i])
```
